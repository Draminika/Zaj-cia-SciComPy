```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, iter):
    z = 0
    for i in range(iter):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return iter

width = 500
height = 500
xmin, xmax = -2.5, 1.0
ymin, ymax = -2, 2
iter = 1500

img = np.zeros((height, width))

for x in range(width):
    for y in range(height):
        real = xmin + (xmax - xmin) * x / width
        imag = ymin + (ymax - ymin) * y / height
        c = real + 1j * imag
        img[y, x] = mandelbrot(c, iter)

plt.imshow(img, cmap='hot', extent=(xmin, xmax, ymin, ymax))
plt.title("Mandelbrot Set")
plt.xlabel("Real(c)")
plt.ylabel("Imaginary(c)")
plt.show()
```
